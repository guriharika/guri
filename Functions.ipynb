**Q6.Explain the concept of generators in Python and how they are defined.**

Ans:
**Generators in Python**

A generator is a special type of iterator in Python that is defined using a function and the yield keyword instead of return. Generators allow for lazy evaluation, meaning they produce values one at a time and only when required, which is more memory-efficient for large datasets.

**How Generators are Defined**

Using the yield Keyword:

* A function that contains yield is called a generator function.

* When the generator function is called, it returns a generator object, not the values themselves.

* Each time yield is encountered, the function returns the yielded value and pauses its state, resuming from where it left off on the next next() call.

**Example:**

def count_up_to_three():

    yield 1
    yield 2
    yield 3

#Create the generator

counter = count_up_to_three()

#Get values from the generator

print(next(counter))  # Output: 1

print(next(counter))  # Output: 2

print(next(counter))  # Output: 3

**Q7.What are the advantages of using generators over regular functions?**

Ans:

**Memory Efficiency:**

Generators produce values one at a time, so they use less memory compared to returning entire collections (e.g., lists).

**Example:**

####def count_up_to(max):

    count = 1

    while count <= max:

        yield count

        count += 1

gen = count_up_to(1000000)  # Uses less memory


**Lazy Evaluation:**

* Values are generated only when needed, which is more efficient for large or infinite datasets.

**Example:**

####def squares():

    for i in range(1, 4):
        yield i ** 2

gen = squares()  # Values are generated one by one


**Performance:**

* Ideal for processing large datasets or streams without loading everything into memory.

**Example**

####def read_file_line_by_line(file):

    for line in file:
        yield line

with open('large_file.txt') as file:

    for line in read_file_line_by_line(file):
        pass  # Reads lines one by one, no memory overload


**Cleaner Code:**

* Generators make the code concise and readable, especially for processing sequences.

**Example**

####def double_numbers(numbers):

    for number in numbers:
        yield number * 2

numbers = [1, 2, 3]

result = list(double_numbers(numbers))  # Output: [2, 4, 6]


**State Retention:**

* Generators maintain their state between iterations, allowing them to resume from where they left off.

**Example**

####def count_up():

    num = 1
    while True:
        yield num
        num += 1

gen = count_up()

print(next(gen))  # Output: 1

print(next(gen))  # Output: 2 (continues from where it left off)


**Infinite Sequences:**

* Generators can handle infinite sequences (e.g., endless counting) without consuming infinite memory.

**Example**

####def fibonacci():

    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()

for _ in range(5):

    print(next(fib))  # Output: 0, 1, 1, 2, 3

**Q10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

Ans:

**map():**

* Purpose: Applies a given function to each item in an iterable and returns an iterator that produces the transformed results.

* Usage: When you want to apply a function to all elements in an iterable.

**Example**

numbers = [1, 2, 3, 4]

squared_numbers = map(lambda x: x ** 2, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16]

**reduce():**

* Purpose: Applies a given function cumulatively to the items in an iterable, reducing the iterable to a single value.

* Usage: When you want to perform a rolling computation on an iterable and accumulate the result.

**Example:**

from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, numbers)

print(result)  # Output: 10


**filter():**

* Purpose: Filters elements from an iterable based on a condition defined by a function and returns an iterator with only the elements that satisfy the condition.

* Usage: When you want to filter out elements from an iterable based on a certain condition.

**Example:**

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = filter(lambda x: x % 2 == 0, numbers)

print(list(even_numbers))  # Output: [2, 4, 6]

```python
def sum_even_numbers(*a):
  n=[]
  for i in a:
    if type(i)==int and i%2==0:
      n.append(i)
  return n
```

**Q4.Write a Python function that checks if a given number is prime or not from 1 to 200**

Ans:

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
is_prime(2)
```

```python
is_prime(7)
```

**Q11.) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:**

**Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100,00 .**

**Write a Python program using lambda and map.**

Ans:



```python
books = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]
result = list(map(lambda book: (
    book[0],
    round(book[2] * book[3] * 1.1, 2) if book[2] * book[3] < 100 else round(book[2] * book[3], 2)
), books))
```

```python
result
```

book[0]:

* Refers to the order number of the book.

book[2]:

* Refers to the quantity of books.

book[3]:

* Refers to the price per item of the book.

book[2] * book[3]:

* Calculates the total price (quantity  price per item).

if book[2] * book[3] < 100:

* Checks if the total price is less than 100.

round(book[2] * book[3] * 1.1, 2):

* If the total price is less than 100, it increases it by 10% (multiply by 1.1) and rounds the result to 2 decimal places.

else round(book[2] * book[3], 2):

* If the total price is 100 or greater, it simply rounds the total price to 2 decimal places.

**Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis**

Ans:The null hypothesis and the alternative hypothesis are two complementary statements used in hypothesis testing. They help researchers test a claim or assumption about a population based on sample data.

1.Null Hypothesis (H₀)
The null hypothesis represents the default assumption or the "status quo." It suggests that there is no effect, no difference, or no relationship between the variables being tested. It's essentially saying that any observed effect in the data is due to chance or random variation.
It is often a statement of no difference or no association.

**Key points about the Null Hypothesis (H₀):**

* It is what you are trying to test against.

* It assumes that the results of an experiment are due to random chance unless proven otherwise.

* It is often phrased as: “There is no effect,” “There is no difference,” or “The variables are independent.”

**Q3.What is the significance level in hypothesis testing, and why is it important**


Ans:1.
**The significance level (α)**is the probability of rejecting the null hypothesis when it is actually true. It is the threshold that defines how much risk of a Type I error (false positive) you are willing to accept in your test.

 * α = 0.05 is the most common choice in many scientific fields. This means that you are willing to accept a 5% chance of wrongly rejecting the null hypothesis when it is actually true.

* You can also use α = 0.01 (1% risk) or α = 0.10 (10% risk), depending on how strict you want to be about avoiding Type I errors.

**Why is the Significance Level Important?**

The significance level is important because it helps you decide whether your results are statistically significant. Here’s why it matters:

* Defining the Risk of Type I Error: The significance level represents the probability of making a Type I error, which occurs when you reject a true null hypothesis. A lower significance level (like 0.01) reduces the risk of making a false positive error but makes it harder to reject the null hypothesis. A higher significance level (like 0.10) increases the chance of rejecting the null hypothesis but also increases the risk of making a false positive error.

* Threshold for Evidence: The significance level sets the threshold for determining if the observed data is unusual enough to reject the null hypothesis. If the p-value from your hypothesis test is smaller than α, the result is considered statistically significant, and you reject the null hypothesis. If the p-value is greater than α, you fail to reject the null hypothesis.

**Q5.How do you interpret the P-value in hypothesis testing**

Ans:Interpreting the p-value in hypothesis testing is a critical part of the decision-making process. The p-value helps you assess the strength of the evidence against the null hypothesis (H₀) and decide whether to reject or fail to reject it. Here's a clear breakdown of how to interpret it:

**1. P-value ≤ Significance Level (α)**

* What it means: If the p-value is less than or equal to the significance level (α), you reject the null hypothesis.

* Interpretation: The data provides strong evidence against the null hypothesis. You can conclude that the observed effect is statistically significant.

* Example:
Suppose your α = 0.05 and you get a p-value of 0.03.

* Since 0.03 ≤ 0.05, you reject the null hypothesis.
* Conclusion: The result is statistically significant, and there is evidence to support the alternative hypothesis (H₁).



**2. P-value > Significance Level (α)**

* What it means: If the p-value is greater than the significance level (α), you fail to reject the null hypothesis.

* Interpretation: The data does not provide strong enough evidence to reject the null hypothesis. This means the result is not statistically significant.

* Example:
Suppose your α = 0.05 and you get a p-value of 0.08.

* Since 0.08 > 0.05, you fail to reject the null hypothesis.
* Conclusion: The result is not statistically significant, and you cannot conclude that there is a meaningful effect.

**Q6.What are Type 1 and Type 2 errors in hypothesis testing**

Ans:In hypothesis testing, Type I and Type II errors refer to the potential mistakes or misinterpretations that can occur when making decisions about the null hypothesis (H₀). These errors arise when you either reject the null hypothesis when you shouldn't (Type I error) or fail to reject it when you should (Type II error)

**1. Type I Error (False Positive)**
* What is it?
A Type I error occurs when you reject the null hypothesis even though it is actually true. In other words, you conclude that there is a significant effect or relationship when, in reality, there isn't one. This is often called a false positive.

* Example:
In a drug trial, the null hypothesis might be: "The new drug has no effect on recovery time." If you reject this hypothesis based on the data, even though the drug has no effect, you are committing a Type I error. Essentially, you "think" the drug works when it actually doesn't.

**2. Type II Error (False Negative)**
* What is it?
A Type II error occurs when you fail to reject the null hypothesis when it is actually false. In other words, you miss detecting a true effect or difference. This is often called a false negative.

* Example:
Continuing with the drug trial example, if the null hypothesis is "The new drug has no effect on recovery time," and the drug actually does improve recovery time, but your test fails to detect this difference, you are committing a Type II error. Essentially, you "think" the drug doesn't work when it actually does.

**2. Two-Tailed Test**

It is used when you want to check if a parameter (like a mean or proportion) is significantly different from the hypothesized value, without specifying the direction of the difference. This test is "non-directional," meaning you are open to the possibility of either an increase or a decrease.

**Q11.What is the difference between a Z-test and a T-test**

Ans:1. Population Standard Deviation (σ) vs. Sample Standard Deviation (s)

* Z-test:
The Z-test is used when the population standard deviation (σ) is known. In most practical scenarios, this is rare, but when the population standard deviation is available or can be accurately estimated, the Z-test is the appropriate choice.

* T-test:
The T-test is used when the population standard deviation is unknown, and you are using the sample standard deviation (s) to estimate the population parameter. This happens most often in practice, especially when dealing with small sample sizes.

2. Sample Size (n)
* Z-test:
The Z-test is typically used when the sample size is large (n ≥ 30). According to the Central Limit Theorem, for large sample sizes, the sampling distribution of the sample mean will be approximately normal, even if the population distribution is not normal. Hence, the Z-test can be used effectively when sample sizes are large.

* T-test:
The T-test is used when the sample size is small (n < 30). With smaller samples, the sample mean's distribution is more variable, and the T-distribution (which has heavier tails) helps account for that additional uncertainty.

3. Hypothesis Testing
Both tests are used for hypothesis testing (such as one-sample, two-sample tests, or paired-sample tests). The primary difference is in the data assumptions and the test statistic used.

* Z-test:
When performing a Z-test, you compare the Z-statistic to a critical value from the standard normal distribution. If the absolute value of the Z-statistic exceeds the critical value, you reject the null hypothesis.

* T-test:
In a T-test, you compare the T-statistic (which takes into account sample size and variability) to a critical value from the T-distribution. The T-statistic follows the T-distribution with degrees of freedom (typically df = n - 1).

**Q14.What is a confidence interval, and how is it used to interpret statistical results**

Ans:A confidence interval (CI) is a range of values that is used to estimate a population parameter (such as a population mean or proportion) with a certain level of confidence. It provides a range of plausible values for the parameter, given the sample data, and quantifies the uncertainty around the estimate.

Interpreting Confidence Intervals:
The interpretation of a confidence interval depends on the confidence level and the context of the problem. Here’s how to interpret the results:

1. Range of Possible Values:
A 95% confidence interval means that we are 95% confident that the true population parameter lies within the interval. For example, if we have a 95% CI for a population mean, and the interval is [48, 52], we can say that we are 95% confident that the true population mean lies somewhere between 48 and 52.

2. Precision of the Estimate:
The width of the confidence interval indicates the precision of the estimate. A narrow interval means the estimate is more precise, while a wide interval means there is more uncertainty or variability in the estimate.

* Narrow CI: Indicates that the sample data provide a relatively precise estimate of the population parameter.
* Wide CI: Indicates that the sample data have more variability, or the sample size is too small, leading to less precision.

**Q21.What are the different types of ANOVA tests**

Ans:**1. One-Way ANOVA (also called One-Factor ANOVA)**
Purpose: Compares the means of three or more independent groups based on one factor (or independent variable).

Assumptions:

* The groups are independent of each other.
* The dependent variable is continuous.
*  The data in each group follow a normal distribution.
* Homogeneity of variances (the variance within each group is roughly equal).
* Example: Testing the mean test scores of students who used different study methods (e.g., method A, method B, method C).

**2. Two-Way ANOVA (also called Two-Factor ANOVA)**

Purpose: Compares the means of groups that have been split based on two factors (independent variables). This type of ANOVA can also test if there is an interaction effect between the two factors.

Two-way ANOVA with interaction: Tests not only the main effects of each factor, but also if there is an interaction effect between the two factors.

Two-way ANOVA without interaction: Tests the main effects of each factor separately, assuming there is no interaction between them.

Assumptions:

* The factors are independent.
* Each combination of factor levels is independent.
* The dependent variable is continuous.
* Normality of data within each group.
* Homogeneity of variances for all groups.
8 Example: Studying the effects of both diet type (e.g., low-carb, high-carb) and exercise level (e.g., low, medium, high) on weight loss. You want to know if there is an interaction between diet and exercise on weight loss.

**3. Repeated Measures ANOVA**

Purpose: Used when the same subjects are measured multiple times under different conditions or at different time points. This is particularly useful for within-subjects designs, where each participant is exposed to all levels of the independent variable.

Assumptions:

* The dependent variable is continuous.
* The data are correlated within subjects (since the same subjects are measured repeatedly).
* The sphericity assumption (the variances of the differences between levels of the independent variable are equal) must be met. If violated,
Greenhouse-Geisser correction may be applied.
* Example: Testing the effects of different teaching methods on student performance at three time points: before the course, during the course, and after the course. Each student is measured at all three time points.

F-test and Its Relationship to Hypothesis Testing

**The general form of the F-test hypothesis is:**

* Null hypothesis (H₀): There is no significant difference between the variances (or effects) in the groups being compared.

* Alternative hypothesis (H₁): There is a significant difference between the variances (or effects).
* In ANOVA: H₁ assumes that at least one group mean differs from the others.
* In regression: H₁ suggests that at least one of the independent variables explains a significant amount of the variance in the dependent variable.

**Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results**

Ans:

```python
import numpy as np
from scipy import stats

# Sample data (for example, heights of students in a class)
sample_data = [170, 175, 160, 165, 180, 170, 168, 172, 178, 165]

# Known population mean (e.g., population mean height of students in a school)
population_mean = 170

# Known population standard deviation (e.g., population std deviation of student height)
population_std = 6

# Sample size
n = len(sample_data)

# Sample mean
sample_mean = np.mean(sample_data)

# Z-test formula: Z = (sample_mean - population_mean) / (population_std / sqrt(n))
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Perform the two-tailed Z-test (alternative hypothesis: sample mean != population mean)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Display results
print(f"Sample Mean: {sample_mean}")
print(f"Population Mean: {population_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpret the results based on a significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

**Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate random data (e.g., heights of 30 students)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=170, scale=5, size=30)

# Step 2: Define population parameters
population_mean = 172  # Hypothetical population mean
population_std = 5     # Known population standard deviation

# Step 3: Calculate sample mean and sample size
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 4: Calculate the Z-statistic for a two-tailed test
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: Calculate the p-value for a two-tailed test
# For a two-tailed test, we calculate the p-value as 2 * (1 - CDF of the absolute value of Z)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Step 6: Define the significance level and calculate the critical Z-values
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)  # For two-tailed test (alpha/2 for each tail)

# Step 7: Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Population Mean: {population_mean}")
print(f"Z-statistic: {z_statistic:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 8: Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")

# Step 9: Visualization of Z-distribution and decision regions

# Plot the standard normal distribution (Z-distribution)
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y, label="Standard Normal Distribution")
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (Right)')
plt.axvline(x=z_statistic, color='blue', linestyle='dashed', label=f"Z-statistic = {z_statistic:.3f}")

# Mark the critical values
plt.axvline(x=-z_critical, color='green', linestyle='dashed', label=f"Critical Z = {-z_critical:.3f}")
plt.axvline(x=z_critical, color='green', linestyle='dashed', label=f"Critical Z = {z_critical:.3f}")

# Labeling and plot decoration
plt.title("Two-Tailed Z-Test: Decision Region")
plt.xlabel("Z-values")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

**Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing**

Ans:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def visualize_errors(sample_size, population_mean, population_std, alternative_mean, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors during hypothesis testing.

    Parameters:
    - sample_size: Size of the sample.
    - population_mean: Mean under the null hypothesis (H₀).
    - population_std: Population standard deviation.
    - alternative_mean: Mean under the alternative hypothesis (H₁).
    - alpha: Significance level (default is 0.05).
    """

    # Step 1: Define critical value for Z-test
    z_critical = stats.norm.ppf(1 - alpha)  # Critical Z-value for two-tailed test

    # Step 2: Calculate the probability of Type 1 error (Rejecting H₀ when it's true)
    # Type 1 error occurs when the sample mean falls in the rejection region (beyond z_critical)
    type_1_error_area = stats.norm.cdf(-z_critical)  # Left tail

    # Step 3: Calculate the probability of Type 2 error (Failing to reject H₀ when it's false)
    # Type 2 error occurs when the sample mean falls in the non-rejection region, but H₁ is true
    z_beta = (z_critical - (alternative_mean - population_mean)) / (population_std / np.sqrt(sample_size))
    type_2_error_area = stats.norm.cdf(z_beta)  # Left tail of the distribution under H₁

    # Step 4: Calculate Power of the test (1 - Type 2 Error)
    power = 1 - type_2_error_area

    # Display results
    print(f"Type 1 Error (α): {type_1_error_area:.4f}")
    print(f"Type 2 Error (β): {type_2_error_area:.4f}")
    print(f"Power of the test (1 - β): {power:.4f}")

    # Step 5: Visualize the errors and decision regions
    x = np.linspace(population_mean - 5 * population_std, population_mean + 5 * population_std, 1000)
    y = stats.norm.pdf(x, loc=population_mean, scale=population_std / np.sqrt(sample_size))

    plt.figure(figsize=(10, 6))

    # Plot the normal distribution for H₀ (null hypothesis) and H₁ (alternative hypothesis)
    plt.plot(x, y, label="Null Hypothesis Distribution (H₀)", color='blue')
    plt.fill_between(x, 0, y, where=(x <= population_mean - z_critical * population_std / np.sqrt(sample_size)),
                     color='red', alpha=0.5, label="Rejection Region (Type 1 Error)")

    # Plot the alternative hypothesis distribution
    y_alternative = stats.norm.pdf(x, loc=alternative_mean, scale=population_std / np.sqrt(sample_size))
    plt.plot(x, y_alternative, label="Alternative Hypothesis Distribution (H₁)", color='green', linestyle='dashed')
    plt.fill_between(x, 0, y_alternative, where=(x >= population_mean + z_critical * population_std / np.sqrt(sample_size)),
                     color='orange', alpha=0.5, label="Non-Rejection Region (Type 2 Error)")

    # Mark critical regions and decision boundaries
    plt.axvline(x=population_mean + z_critical * population_std / np.sqrt(sample_size), color='green', linestyle='dashed', label=f"Critical Value (Z = {z_critical:.2f})")

    plt.title("Visualizing Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Usage:
# Let's assume we are comparing the sample mean of students' heights to a population mean.
# H₀: population mean = 172, H₁: alternative mean = 170
sample_size = 30
population_mean = 172  # Under H₀
population_std = 5     # Standard deviation of the population
alternative_mean = 170  # The mean under the alternative hypothesis (H₁)
alpha = 0.05           # Significance level

# Call the function to calculate and visualize Type 1 and Type 2 errors
visualize_errors(sample_size, population_mean, population_std, alternative_mean, alpha)
```

**Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance**

Ans:

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - sample_data: The sample data (array-like).
    - confidence_level: The desired confidence level (default is 0.95).

    Returns:
    - confidence_interval: The lower and upper bounds of the confidence interval.
    """

    # Step 1: Calculate sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (Bessel's correction)
    sample_size = len(sample_data)

    # Step 2: Determine the Z-value (for large samples, assume normal distribution)
    # If the sample size is large, use Z-distribution. Otherwise, use t-distribution.
    if sample_size > 30:
        # Z-distribution
        z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = z_value * (sample_std / np.sqrt(sample_size))
    else:
        # T-distribution (used for small samples)
        t_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)
        margin_of_error = t_value * (sample_std / np.sqrt(sample_size))

    # Step 3: Calculate the confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    return confidence_interval

# Example Usage:
# Simulate some data (e.g., test scores of 50 students)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=80, scale=10, size=50)  # Mean = 80, Standard deviation = 10

# Calculate the 95% confidence interval for the sample mean
confidence_level = 0.95
ci = confidence_interval(sample_data, confidence_level)

print(f"Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")
```

**Q11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process**

Ans:

```python
def bayesian_inference(prior_h, likelihood_d_given_h, likelihood_d_given_not_h, prior_not_h):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
    - prior_h: The prior probability of hypothesis (having the disease).
    - likelihood_d_given_h: The likelihood of data given the hypothesis (test positive given the disease).
    - likelihood_d_given_not_h: The likelihood of data given the negation of the hypothesis (test positive given no disease).
    - prior_not_h: The prior probability of the negation of the hypothesis (no disease).

    Returns:
    - posterior_h_given_d: The posterior probability (probability of disease given positive test).
    """

    # Calculate P(D) (the total probability of testing positive)
    p_d = likelihood_d_given_h * prior_h + likelihood_d_given_not_h * prior_not_h

    # Apply Bayes' Theorem
    posterior_h_given_d = (likelihood_d_given_h * prior_h) / p_d

    return posterior_h_given_d

# Example values
prior_h = 0.01  # Prior probability of having the disease (1% prevalence)
likelihood_d_given_h = 0.9  # Sensitivity (probability of testing positive given having the disease)
likelihood_d_given_not_h = 0.05  # False positive rate (probability of testing positive without the disease)
prior_not_h = 1 - prior_h  # Prior probability of not having the disease

# Calculate the posterior probability (probability of having the disease given a positive test result)
posterior = bayesian_inference(prior_h, likelihood_d_given_h, likelihood_d_given_not_h, prior_not_h)

print(f"Posterior Probability of having the disease given a positive test result: {posterior:.4f}")
```

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    - observed_data: A 2D numpy array representing the observed frequencies.

    Returns:
    - expected_frequencies: A 2D numpy array representing the expected frequencies.
    """
    # Calculate the row totals and column totals
    row_totals = np.sum(observed_data, axis=1)  # Sum of each row
    column_totals = np.sum(observed_data, axis=0)  # Sum of each column

    # Calculate the total number of observations
    total = np.sum(observed_data)  # Grand total of the table

    # Initialize an empty array to store expected frequencies
    expected_frequencies = np.zeros_like(observed_data, dtype=float)

    # Calculate expected frequencies for each cell
    for i in range(observed_data.shape[0]):  # Loop through rows
        for j in range(observed_data.shape[1]):  # Loop through columns
            expected_frequencies[i, j] = (row_totals[i] * column_totals[j]) / total

    return expected_frequencies

# Example contingency table (observed data)
observed_data = np.array([[30, 10, 15],  # Male preferences
                          [25, 30, 10]])  # Female preferences

# Calculate expected frequencies
expected_data = calculate_expected_frequencies(observed_data)

# Output the expected frequencies
print("Expected Frequencies:")
print(expected_data)
```

```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1: First sample (array-like).
    - sample2: Second sample (array-like).

    Returns:
    - f_statistic: The calculated F-statistic.
    - p_value: The p-value for the F-test.
    """
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)  # Sample variance (Bessel's correction)
    var2 = np.var(sample2, ddof=1)  # Sample variance (Bessel's correction)

    # Calculate the F-statistic (larger variance / smaller variance)
    if var1 > var2:
        f_statistic = var1 / var2
    else:
        f_statistic = var2 / var1

    # Degrees of freedom for each sample
    df1 = len(sample1) - 1  # df for sample1
    df2 = len(sample2) - 1  # df for sample2

    # Calculate the p-value using the F-distribution
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    return f_statistic, p_value

# Example: Generate two random samples with different variances
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=10, scale=5, size=50)  # Sample 1 with a standard deviation of 5
sample2 = np.random
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check the assumptions for One-Way ANOVA
def check_anova_assumptions(*groups):
    # 1. Check for Normality: Shapiro-Wilk test for normality
    print("Checking normality using the Shapiro-Wilk test:")
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: Shapiro-Wilk test statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i} fails normality assumption (p < 0.05).")
        else:
            print(f"Group {i} passes normality assumption (p >= 0.05).")

    # 2. Check for Equal Variance: Levene's test
    print("\nChecking equal variance assumption using Levene's test:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("Groups fail the equal variance assumption (p < 0.05).")
    else:
        print("Groups pass the equal variance assumption (p >= 0.05).")

    # 3. Visualizations for normality and independence
    print("\nVisualizing the data:")

    # Boxplot to visualize the spread and identify outliers
    plt.figure(figsize=(8, 6))
    plt.boxplot(groups, labels=[f'Group {i+1}' for i in range(len(groups))])
    plt.title('Boxplot of Groups')
    plt.ylabel('Values')
    plt.grid(True)
    plt.show()

    # Q-Q plot to check for normality
    for i, group in enumerate(groups, 1):
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Q-Q plot for Group {i}")
        plt.show()

    # 4. Check for independence (optional): If you have time series data, you can use Durbin-Watson test
    # This step can be skipped for most situations unless time series data is being used.

# Example groups with test scores for three different study methods
group1 = np.array([85, 90, 88, 92, 87])  # Group 1: Studied method A
group2 = np.array([78, 82, 85, 80, 79])  # Group 2: Studied method B
group3 = np.array([93, 95, 97, 94, 96])  # Group 3: Studied method C

# Call the function to check assumptions
check_anova_assumptions(group1, group2, group3)
```

**Q21.Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing**

Ans:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Function to visualize the F-distribution
def plot_f_distribution(df1, df2):
    # Create an array of x values from 0 to 5, as the F-distribution is right-skewed
    x = np.linspace(0, 5, 500)

    # Get the probability density function (PDF) of the F-distribution
    y = f.pdf(x, df1, df2)

    # Plot the F-distribution
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'F-distribution: df1={df1}, df2={df2}', color='blue')
    plt.fill_between(x, y, color='skyblue', alpha=0.4)
    plt.title(f'F-Distribution with df1={df1} and df2={df2}')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example: Visualize the F-distribution for two different sets of degrees of freedom
df1 = 5  # Degrees of freedom for the numerator (between-group variance)
df2 = 10  # Degrees of freedom for the denominator (within-group variance)

# Plot the F-distribution
plot_f_distribution(df1, df2)

# Additional Example: Visualize for different degrees of freedom
df1_alt = 3
df2_alt = 20
plot_f_distribution(df1_alt, df2_alt)
```

**Q22.Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means**

Ans:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate synthetic data for three study methods
np.random.seed(42)

# Group 1: Study Method A
group1 = np.random.normal(75, 5, 30)  # Mean = 75, SD = 5, Sample size = 30

# Group 2: Study Method B
group2 = np.random.normal(78, 6, 30)  # Mean = 78, SD = 6, Sample size = 30

# Group 3: Study Method C
group3 = np.random.normal(82, 7, 30)  # Mean = 82, SD = 7, Sample size = 30

# Step 2: Perform a One-Way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Step 3: Output the results of the ANOVA test
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between the means of the groups.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the means of the groups.")

# Step 4: Visualization - Boxplot to visualize the distribution of each group
plt.figure(figsize=(8, 6))
sns.boxplot(data=[group1, group2, group3], labels=['Method A', 'Method B', 'Method C'])
plt.title('Comparison of Test Scores by Study Method')
plt.xlabel('Study Method')
plt.ylabel('Test Scores')
plt.grid(True)
plt.show()
```

**Q23.D Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means**

Ans:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)
group1 = np.random.normal(85, 8, 30)  # Mean=85, SD=8, Sample size=30
group2 = np.random.normal(90, 8, 30)  # Mean=90, SD=8, Sample size=30

# Perform t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

# Output results
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")

# Conclusion
if p_value < 0.05:
    print("Reject null hypothesis: Means are significantly different.")
else:
    print("Fail to reject null hypothesis: No significant difference.")

# Visualization
plt.boxplot([group1, group2], labels=['Group 1', 'Group 2'])
plt.title('Comparison of Means')
plt.show()
```

**Q24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results**

Ans:

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Mean=100, SD=15, Sample size=50

# Step 2: Calculate sample variance
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for unbiased estimator)

# Step 3: Set hypothesized population variance
population_variance = 225  # Hypothesized variance (SD^2 = 15^2 = 225)

# Step 4: Perform Chi-square test for variance
n = len(sample_data)  # Sample size
chi_square_stat = (n - 1) * sample_variance / population_variance

# Step 5: Calculate p-value
df = n - 1  # Degrees of freedom
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Step 6: Output the results
print(f"Chi-square Statistic: {chi_square_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample variance is significantly different from the hypothesized population variance.")
else:
    print("Fail to reject the null hypothesis: The sample variance is not significantly different from the hypothesized population variance.")
```

**Q25.D Write a Python script to perform a Z-test for comparing proportions between two datasets or groups**

Ans:

```python
import numpy as np
import scipy.stats as stats

# Step 1: Simulate data for two groups
# Group 1: 100 successes out of 500 trials
group1_successes = 100
group1_trials = 500

# Group 2: 120 successes out of 600 trials
group2_successes = 120
group2_trials = 600

# Step 2: Calculate sample proportions
p1_hat = group1_successes / group1_trials
p2_hat = group2_successes / group2_trials

# Step 3: Calculate pooled proportion
p_pool = (group1_successes + group2_successes) / (group1_trials + group2_trials)

# Step 4: Calculate the Z-test statistic
z_stat = (p1_hat - p2_hat) / np.sqrt(p_pool * (1 - p_pool) * (1/group1_trials + 1/group2_trials))

# Step 5: Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

# Step 6: Output results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```

**Q26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results**

Ans:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data for two groups
# Group 1: Mean = 50, SD = 10, Sample size = 30
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)

# Group 2: Mean = 55, SD = 20, Sample size = 30
group2 = np.random.normal(55, 20, 30)

# Step 2: Calculate the sample variances
var1 = np.var(group1, ddof=1)  # Sample variance for group 1
var2 = np.var(group2, ddof=1)  # Sample variance for group 2

# Step 3: Perform the F-test
f_stat = var1 / var2  # F-statistic is the ratio of variances
df1 = len(group1) - 1  # Degrees of freedom for group 1
df2 = len(group2) - 1  # Degrees of freedom for group 2

# Step 4: Calculate the p-value (two-tailed)
p_value = 2 * min(stats.f.cdf(f_stat, df1, df2), 1 - stats.f.cdf(f_stat, df1, df2))

# Step 5: Output results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Step 7: Visualization (Boxplot)
plt.figure(figsize=(8, 6))
plt.boxplot([group1, group2], labels=['Group 1', 'Group 2'])
plt.title('Comparison of Variances Between Group 1 and Group 2')
plt.ylabel('Values')
plt.grid(True)
plt.show()

# Visualize the distributions of the two groups with histograms
plt.figure(figsize=(8, 6))
plt.hist(group1, bins=15, alpha=0.5, label='Group 1', color='blue')
plt.hist(group2, bins=15, alpha=0.5, label='Group 2', color='green')
plt.title('Histogram of Group 1 and Group 2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results**

Ans:

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate observed data (rolling a die 100 times)
np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=100, p=[1/6]*6)

# Step 2: Calculate the observed frequencies
observed_freq = np.array([np.sum(observed_data == i) for i in range(1, 7)])

# Step 3: Set up the expected frequencies (uniform distribution for a fair die)
expected_freq = np.array([100 / 6] * 6)  # 100 rolls, each face should appear 100/6 times

# Step 4: Perform the Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

# Step 5: Output the results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Interpretation of the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The die is biased (observed frequencies differ from expected).")
else:
    print("Fail to reject the null hypothesis: The die is fair (no significant difference).")

# Step 7: Visualization of observed vs expected frequencies
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))

# Create a bar plot for observed vs expected frequencies
plt.figure(figsize=(8, 6))
plt.bar(x - 0.2, observed_freq, 0.4, label='Observed', color='blue')
plt.bar(x + 0.2, expected_freq, 0.4, label='Expected', color='orange')
plt.xticks(x, labels)
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies for Die Rolls')
plt.legend()
plt.show()
```
